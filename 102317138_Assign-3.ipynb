```python
average_salary = np.mean(csvdata['Salary'])
bonus = np.sum(csvdata['Bonus'])
emp_age = np.min(csvdata['Age'])
rate = np.max(csvdata['Rating'])
print(average_salary)
print(bonus)
print(emp_age)
print(rate)
```

```python
csvdata.sort_values(by='Salary',ascending=False)
```

```python
def categorize_performance(rating):
  if rating >= 4.5:
    return 'Excellent'
  elif rating >= 4.0:
    return 'Good'
  else:
    return 'Average'

csvdata['Performance Category'] = csvdata['Rating'].apply(categorize_performance)
csvdata
```

```python
def tax_red(salary):
  return (salary*0.1)

csvdata['Tax'] = csvdata['Salary'].apply(tax_red)
csvdata
```
